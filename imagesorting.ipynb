```python
from cs.src.computervision import MODEL_CFG_DIR, MODEL_WEIGTS_DIR_DRACO, CLASSES, ObjectDetection
import cv2
import numpy as np

net = ObjectDetection(MODEL_WEIGTS_DIR_DRACO)

with open("success_max_res.txt", "r") as f:
    hits = f.read().split("\n")

def test_image(filedir: str) -> bool:
    img = cv2.imread(filedir)

    leaders = net.process(img)

    if len(leaders) > 0:
        img = leaders[0].render(img)

        if filedir in hits:
            cv2.imshow('Image', img)
            cv2.waitKey(0)
            cv2.destroyAllWindows()

            if input("Is valid image: ") == "y":
                return 1

        return 3

    return 2
```

```python
images_dir = "C:\\Users\\alvar\\University of Southampton\\GDP - General\\Flight Day 2\\images"

from os import listdir
from os.path import isfile, join
images = [join(images_dir, f) for f in listdir(images_dir) if isfile(join(images_dir, f))]



for i in images:
    if i.split("\\")[-1] == "0.jpg" and not test_image(i):
        raise EnvironmentError("Test image failed")
    else:
        print(i)
        try:
            output = test_image(i)
            if output == 1:
                with open("success.txt", "a") as f:
                    f.write(f"{i}\n")

            elif output == 2:
                with open("fail.txt", "a") as f:
                    f.write(f"{i}\n")
            else:
                with open("incorrect.txt", "a") as f:
                    f.write(f"{i}\n")
        except Exception as e:
            print(e)
```
